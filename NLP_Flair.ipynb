```python
import flair
```

```python
#Distill BERT
classification_model = flair.models.TextClassifier.load('en-sentiment')
```

```python
classification_model
```

```python
text = "You are great"  
sentence = flair.data.Sentence(text)
sentence
```

```python
type(sentence)
```

```python
sentence.to_tokenized_string()
```

```python
classification_model.predict(sentence)
```

```python
print(sentence)
print(sentence.get_labels())
print(sentence.get_labels()[0])
print(sentence.get_labels()[0].score)
print(sentence.get_labels()[0].value)
```
